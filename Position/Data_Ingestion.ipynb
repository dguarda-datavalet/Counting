```python
#saving bigtable
import pandas_gbq
from datalab.context import Context
import datalab.storage as storage
import datalab.bigquery as bq
from pandas import DataFrame
import time
from tqdm import tqdm
tqdm.pandas()
```

```python
%%bigquery df_onroute_position_raw
SELECT scout.device_mac, scout.ap_oem, scout.channel, scout.Timestamp_ms, scout.Timestamp, datetime(timestamp_seconds(scout.timestamp)) as local_time, scout.is_associated,
scout.signal_strength, scout.noise_floor, scout.channel, hard.brand_name, hard.market_segment, scout.ap_mac, hard.site_uuid
FROM  data-prod-270222.raw.rtls_scout_raw scout
join meta.hardware_details as hard on utils.normalize_mac(scout.ap_mac) = utils.normalize_mac(hard.mac)
where scout.timestamp between 1637333100 and 1637336700
and hard.site_uuid = '657b300b-db2e-f810-ffc6-e3a84b96192c'
order by scout.timestamp
limit 100000
```

```python
start = time.time()
df_onroute_position_raw.to_gbq('data-prod-270222.datascience.position_raw', 
                 'data-prod-270222',
                 chunksize=10000, 
                 if_exists='replace'
                 )
end = time.time()
print("time alternative 1 " + str(end - start))
```
